### Diabetes Prediction using Kmeans
Le diabète est une maladie chronique courante et constitue une grande menace pour la santé humaine. La caractéristique du diabète est que la glycémie est supérieure au niveau normal, ce qui est causé par une sécrétion défectueuse d'insuline ou par ses effets biologiques altérés, ou les deux. Le diabète peut entraîner des lésions chroniques et un dysfonctionnement de divers tissus, en particulier des yeux, des reins, du cœur, des vaisseaux sanguins et des nerfs. Le diabète peut être divisé en deux catégories, le diabète de type 1 (DT1) et le diabète de type 2 (DT2). Les patients atteints de diabète de type 1 sont normalement plus jeunes, généralement âgés de moins de 30 ans. Les symptômes cliniques typiques sont une augmentation de la soif et des mictions fréquentes, une glycémie élevée. Ce type de diabète ne peut pas être guéri efficacement avec des médicaments oraux seuls et les patients doivent recevoir une insulinothérapie. Le diabète de type 2 survient plus fréquemment chez les personnes d'âge moyen et les personnes âgées, ce qui est souvent associé à l'apparition d'obésité, d'hypertension, de dyslipidémie, d'artériosclérose et d'autres maladies.

### KMEANS CLUSTERING
Le cluster est défini comme des groupes de points de données tels que les points de données d'un groupe seront similaires ou liés les uns aux autres et différents des points de données d'un autre groupe. Et le processus est connu sous le nom de regroupement. L'objectif du clustering est de déterminer le groupement intrinsèque dans un ensemble de données non étiquetées. Le clustering est une forme d'apprentissage non supervisé car il ne nécessite pas de données étiquetées.

K-means est un algorithme de clustering partitionnel non supervisé qui est basé sur le regroupement de données en k - nombre de clusters en déterminant le centroïde à l'aide de la méthode euclidienne ou Manhattan pour le calcul de la distance. Il regroupe l'objet en fonction de la distance minimale.

TROUVER LE MEILLEUR K PAR LA MÉTHODE DU COUDE
Tout en travaillant avec le zéro de clustering k-means, une chose que nous devons garder à l'esprit est le nombre de clusters 'k'. Nous devons nous assurer que nous choisissons le nombre optimal de clusters pour l'ensemble de données donné. Mais, ici se pose une question, comment choisir la valeur optimale de k ?? Nous utilisons la méthode du coude qui est généralement utilisée pour analyser la valeur optimale de k.

La méthode Elbow est basée sur le principe selon lequel "la somme des carrés des distances de chaque point de données à partir de son centre de gravité de cluster correspondant doit être aussi minimale que possible".

ÉTAPES POUR CHOISIR LA MEILLEURE VALEUR K
1. Exécutez le modèle de clustering k-means sur différentes valeurs de k
2. Pour chaque valeur de K, calculez la somme des carrés des distances de chaque point de données à partir de son centroïde de cluster correspondant, appelé WCSS (sommes des carrés intra-cluster).
3. Tracer la valeur de WCSS par rapport à diverses valeurs de K
4. Pour sélectionner la valeur de k, nous choisissons la valeur où il y a une courbure (coude) sur le tracé, c'est-à-dire que WCSS n'augmente pas rapidement.
Puisque nous devons regrouper le diabète et le non diabète, cluster = 2 pour notre scénario de problème

### AJUSTEMENT DU MODÈLE (K=2)- KMEANS CLUSTERING

```python
# Selecting features - [Glucose, Insulin, BMI]
X = data1.iloc[:, [1, 4, 5]].values
Y = data1.iloc[:, 8].values
```

```python
# Splitting X and Y
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 42, stratify = BD_diabetes['Outcome'] )
```

```python
# Checking dimensions
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
import sklearn
from sklearn.cluster import KMeans
KMeans_Clustering = KMeans(n_clusters =2, random_state=0)
KMeans_Clustering.fit(X_train)
```

```python
print(KMeans_Clustering.cluster_centers_)
```

```python
#prediction using kmeans and accuracy
kpred = KMeans_Clustering.predict(X_test)

print('Classification report:\n\n', sklearn.metrics.classification_report(Y_test,kpred))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
print("Confusion Matrix :")
outcome_labels = sorted(BD_diabetes.Outcome.unique())
sns.heatmap(
    
    confusion_matrix(Y_test, kpred),
    annot=True,fmt="",
    xticklabels=outcome_labels,
    yticklabels=outcome_labels
)
```

Le code que vous avez fourni utilise la fonction confusion_matrix du module sklearn.metrics pour créer une matrice de confusion pour un modèle de prédiction du diabète.

Une matrice de confusion est un tableau qui résume les performances d'un modèle de classification en comparant les étiquettes réelles (vraies) d'un ensemble de données avec les étiquettes prédites du modèle. La matrice de confusion se compose de quatre valeurs : les vrais positifs (TP), les faux positifs (FP), les vrais négatifs (TN) et les faux négatifs (FN).

La fonction confusion_matrix prend deux arguments : les vrais libellés de l'ensemble de test (Y_test) et les libellés prédits de l'ensemble de test (kpred). Il renvoie une matrice indiquant le nombre de vrais positifs, de faux positifs, de vrais négatifs et de faux négatifs.

La fonction sns.heatmap du module seaborn est ensuite utilisée pour créer une carte thermique de la matrice de confusion. L'argument annot=True ajoute des valeurs numériques à chaque cellule de la carte thermique, et les arguments xticklabels et yticklabels sont utilisés pour étiqueter les lignes et les colonnes de la matrice avec les étiquettes uniques de la variable de résultat (Outcome dans ce cas).
